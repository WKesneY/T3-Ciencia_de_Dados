```python
import pandas as pd

# Caminho seguindo o padrão do projeto
path = "../data/raw/netflix_titles.csv"

df = pd.read_csv(path)

df.head()
```

```python
colunas_tabela = [
    'Nome da Variável',
    'Tipo',
    'Descrição',
    'Unidades',
    'Valores Faltantes',
    '% Valores Faltantes'
]
```

```python
data_netflix = []
total_registros = df.shape[0]

variaveis = [
    ('show_id', 'Textual', 'Identificador único do título na Netflix.', 'N/A'),
    ('type', 'Categórico', 'Tipo do conteúdo (Filme ou Série).', 'N/A'),
    ('title', 'Textual', 'Título do filme ou série.', 'N/A'),
    ('director', 'Textual', 'Diretor(es) da obra.', 'N/A'),
    ('cast', 'Textual', 'Elenco principal da obra.', 'N/A'),
    ('country', 'Textual', 'País ou países de origem.', 'N/A'),
    ('date_added', 'Data', 'Data em que o título foi adicionado à Netflix.', 'Data'),
    ('release_year', 'Numérico Discreto', 'Ano de lançamento do título.', 'Ano'),
    ('rating', 'Categórico', 'Classificação indicativa.', 'N/A'),
    ('duration', 'Textual', 'Duração do conteúdo (minutos ou temporadas).', 'Minutos/Temporadas'),
    ('listed_in', 'Textual', 'Gêneros associados ao título.', 'N/A'),
    ('description', 'Textual', 'Descrição resumida do conteúdo.', 'N/A')
]

for nome, tipo, descricao, unidade in variaveis:
    faltantes = df[nome].isna().sum()
    percentual = (faltantes / total_registros) * 100

    data_netflix.append([
        nome,
        tipo,
        descricao,
        unidade,
        faltantes,
        round(percentual, 2)
    ])
```

```python
df_tabela = pd.DataFrame(data_netflix, columns=colunas_tabela)
df_tabela
```
